# Histogram and other feature plots

This file makes plots for validating samples

```python
# Import local libraries
import save_and_load
import importlib
import numpy as np
import csv_decoder
from matplotlib import pyplot 
```

```python
# Loading VH and HJ samples
importlib.reload(save_and_load)
vh_event_list, vh_mass_list, vh_higgs, vh_weight, vh_image_list, vh_recluster_images = save_and_load.load('vh-hj', name_1='vh', name_2='hj')
```

```python
# Jet and event image
print('VH event image:')
pyplot.imshow(vh_image_list[0][0])
pyplot.show()
pyplot.imshow(vh_image_list[0][1])
pyplot.show()
pyplot.imshow(vh_image_list[0][2])
pyplot.show()
print('HJ event image:')
pyplot.imshow(hj_image_list[0][0])
pyplot.show()
pyplot.imshow(hj_image_list[0][1])
pyplot.show()
pyplot.imshow(hj_image_list[0][2])
pyplot.show()
```

```python
print('VH jet image:')
pyplot.imshow(vh_recluster_images[0][0])
pyplot.show()
pyplot.imshow(vh_recluster_images[0][1])
pyplot.show()
pyplot.imshow(vh_recluster_images[0][2])
pyplot.show()
print('HJ jet image:')
pyplot.imshow(hj_recluster_images[0][0])
pyplot.show()
pyplot.imshow(hj_recluster_images[0][1])
pyplot.show()
pyplot.imshow(hj_recluster_images[0][2])
pyplot.show()
```

```python
# Cluster 0.8 jets
importlib.reload(csv_decoder)
vh_event_list_clustered = csv_decoder.cluster_event(vh_event_list)
hj_event_list_clustered = csv_decoder.cluster_event(hj_event_list)
vh_reclustered, vh_non_higgs_jets = csv_decoder.recluster_event(vh_event_list_clustered, vh_higgs)
hj_reclustered, hj_non_higgs_jets = csv_decoder.recluster_event(hj_event_list_clustered, hj_higgs)
```

```python
# Extracting data for VH
vh_higgs_pt = []
vh_higgs_eta = []
vh_non_higgs_leading_m = []
vh_non_higgs_leading_pt = []
vh_non_higgs_leading_eta = []
for i in range(len(vh_event_list)):
    vh_higgs_pt.append(vh_higgs[i][0])
    vh_higgs_eta.append(vh_higgs[i][1])
    vh_non_higgs_leading_m.append(vh_non_higgs_jets[i].mass)
    vh_non_higgs_leading_pt.append(vh_non_higgs_jets[i].pt)
    vh_non_higgs_leading_eta.append(vh_non_higgs_jets[i].eta)
    
```

```python
# Extracting data for hj
hj_higgs_pt = []
hj_higgs_eta = []
hj_non_higgs_leading_m = []
hj_non_higgs_leading_pt = []
hj_non_higgs_leading_eta = []
for i in range(len(hj_event_list)):
    hj_higgs_pt.append(hj_higgs[i][0])
    hj_higgs_eta.append(hj_higgs[i][1])
    hj_non_higgs_leading_m.append(hj_non_higgs_jets[i].mass)
    hj_non_higgs_leading_pt.append(hj_non_higgs_jets[i].pt)
    hj_non_higgs_leading_eta.append(hj_non_higgs_jets[i].eta)
```

```python
# Higgs jet mass histogram
pyplot.figure(figsize=(6,4))
n, bins, patches = pyplot.hist(vh_mass_list, 100, facecolor='blue', range=(0,200), label='vh higgs jet mass',histtype = 'step')
n, bins, patches = pyplot.hist(hj_mass_list, 100, facecolor='red', range=(0,200), label='hj higgs jet mass',histtype = 'step')
pyplot.legend()
pyplot.show()
```

```python
# Higgs pT histogram
pyplot.figure(figsize=(6,4))
n, bins, patches = pyplot.hist(vh_higgs_pt, 100, facecolor='blue', label='vh higgs jet pT',histtype = 'step')
n, bins, patches = pyplot.hist(hj_higgs_pt, 100, facecolor='red', label='hj higgs jet pT',histtype = 'step')
pyplot.legend()
pyplot.show()
```

```python
# Higgs eta histogram
pyplot.figure(figsize=(10,6))
n, bins, patches = pyplot.hist(vh_higgs_eta, 100, facecolor='blue', label='vh higgs jet eta',histtype = 'step')
n, bins, patches = pyplot.hist(hj_higgs_eta, 100, facecolor='red', label='hj higgs jet eta',histtype = 'step')
pyplot.legend()
pyplot.show()
```

```python
# Leading non-higgs jet mass histogram
pyplot.figure(figsize=(10,6))
n, bins, patches = pyplot.hist(vh_non_higgs_leading_m, 100, facecolor='blue', range=(0,300), label='vh leading non-higgs jet mass',histtype = 'step')
n, bins, patches = pyplot.hist(hj_non_higgs_leading_m, 100, facecolor='red', range=(0,300), label='hj leading non-higgs jet mass',histtype = 'step')
pyplot.legend()
pyplot.show()
```

```python
# Leading non-higgs jet pT histogram
pyplot.figure(figsize=(10,6))
n, bins, patches = pyplot.hist(vh_non_higgs_leading_pt, 100, facecolor='blue', label='vh leading non-higgs jet pT',histtype = 'step')
n, bins, patches = pyplot.hist(hj_non_higgs_leading_pt, 100, facecolor='red', label='hj leading non-higgs jet pT',histtype = 'step')
pyplot.legend()
pyplot.show()
```

```python
# Leading non-higgs jet eta histogram
pyplot.figure(figsize=(10,6))
n, bins, patches = pyplot.hist(vh_non_higgs_leading_eta, 100, facecolor='blue', range=(-3,3), label='vh leading non-higgs jet eta',histtype = 'step')
n, bins, patches = pyplot.hist(hj_non_higgs_leading_eta, 100, facecolor='red', range=(-3,3), label='hj leading non-higgs jet eta',histtype = 'step')
pyplot.legend()
pyplot.show()
```
